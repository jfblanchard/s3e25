# Cross Validation Functions

```python
from sklearn.metrics import median_absolute_error
from sklearn.model_selection import KFold

def run_cv_and_predict(train, test, features, model, seed, verbose=True):
    
    folds = config.folds
    # initialize arrays 
    fold_scores = []
    oof = np.zeros(train.shape[0])
    preds = np.zeros((test.shape[0],folds))

    # setup folding strategy
    skf = KFold(n_splits=folds,random_state = seed,shuffle=True)

    # start cross validation
    cur_fold = 1
    for trn_idx, val_idx in skf.split(train[features], y):

        # split indicies into train and validation
        x_train = train[features].iloc[trn_idx]
        y_train = y.iloc[trn_idx]
        x_valid = train[features].iloc[val_idx]
        y_valid = y.iloc[val_idx]

        # fit model
        model.fit(x_train,y_train)

        # predict on validation set
        fold_preds = model.predict(x_valid)
        #fold_preds = fold_preds.reshape(len(fold_preds)) # sbabwtdt
        oof[val_idx] = fold_preds

        # Compute scores
        fold_score = median_absolute_error(y_valid,fold_preds)  # CHOOSE METRIC HERE
        fold_scores.append(fold_score)
        if verbose:
            print(f'MAE Score, fold {cur_fold}: {fold_score}')

        # predict on test set - store all fold preds (take mode later)
        test_preds = model.predict(test[features])
        #test_preds = test_preds.reshape(len(test_preds))  # shouldn't have to do this.
        preds[:,cur_fold-1] = test_preds  
        cur_fold +=1
    
    # Print mean fold and oof score 
    oof_score = median_absolute_error(y,oof)
    # oof_score = np.sqrt(mean_squared_error(y,oof))
    print(f'MAE score: {np.mean(fold_scores):.5f}, Stdev: {np.std(fold_scores):.5f}, OOF score: {oof_score:.5f}')
    # print(f'RMSE score: {np.mean(scores):.5f}, Stdev: {np.std(scores):.5f}, OOF score: {oof_score:.5f}')

    return (preds,fold_scores,oof_score,oof)
```

## LightGBM with apply sample weights
* Thanks to [this](https://www.kaggle.com/competitions/playground-series-s3e25/discussion/455888) notebook by AmborsM

```python
from lightgbm import LGBMRegressor

preds = []
#results = []
base_weight = 0.01
lower, upper = 0.06, 0.7

kf = KFold(shuffle=True, random_state=1)

for lr in [.15]:
    for mcs in [50]:
        
        score_without, score_with = 0, 0

        for fold, (idx_tr, idx_va) in enumerate(kf.split(train)):
            X_tr = train[features].iloc[idx_tr].copy()
            X_va = train[features].iloc[idx_va].copy()
            y_tr = y[idx_tr]
            y_va = y[idx_va]

            # Fit and score a model without sample weights
            model_lgbm = LGBMRegressor(objective='mae',
                                  learning_rate=lr,
                                  min_child_samples=mcs,
                                  random_state=config.seed)
            model_lgbm.fit(X_tr, y_tr)
            medae = median_absolute_error(y_va, model_lgbm.predict(X_va))
            score_without += medae

            # Set sample weights
            # Samples with extreme absolute errors get weight 0.01
            # Samples near the median absolute error get weight 1.01
            y_tr_ae = np.abs(y_tr - model_lgbm.predict(X_tr)) # array of absolute errors
            sample_weight_tr = base_weight + ((lower < y_tr_ae) & (y_tr_ae < upper))

            # Fit and score a model with sample weights
            model_lgbm.fit(X_tr, y_tr, sample_weight=sample_weight_tr)
            medae = median_absolute_error(y_va, model_lgbm.predict(X_va))
            score_with += medae

            # predict on test
            preds.append(model_lgbm.predict(test[features]))

        print(#f"# Sample_weight improves CV from {score_without / kf.n_splits:.5f}"
              f"# Score with weights applied {score_with / kf.n_splits:.5f} "
              f"with {base_weight=} and {lower=} and {upper=}"
              f" lr: {lr}, msc:{mcs}")
        
        #results.append([lr,mcs,score_with])

# Sample_weight improves CV from 0.52734 to 0.49772
# with base_weight=0.01 and lower=0.06 and upper=0.7

# Default:
# Sample_weight improves CV from 0.54115 to 0.49913 with base_weight=0.01 and lower=0.06 and upper=0.7

# Just curious to see MSE:
# Sample_weight improves CV from 0.64880 to 0.53244 with base_weight=0.01 and lower=0.06 and upper=0.7
# so mse ratio of worseness is about the same with etc and lgbm

# tuning 1: lr and mcs
# Score with weights applied 0.49833 with base_weight=0.01 and lower=0.06 and upper=0.7 lr: 0.1, msc:50
# Low: Score with weights applied 0.49772 with base_weight=0.01 and lower=0.06 and upper=0.7 lr: 0.15, msc:50

# zoom into tuning:
# weights applied 0.49721 with base_weight=0.01 and lower=0.06 and upper=0.7 lr: 0.12, msc:47, lb = .46762
# weights applied 0.49038 with base_weight=0.01 and lower=0.06 and upper=0.7 lr: 0.12, msc:49
# weights applied 0.49121 with base_weight=0.01 and lower=0.06 and upper=0.7 lr: 0.13, msc:53
# weights applied 0.49158 with base_weight=0.01 and lower=0.06 and upper=0.7 lr: 0.15, msc:47, lb = .47273
# weights applied 0.48942 with base_weight=0.01 and lower=0.06 and upper=0.7 lr: 0.15, msc:52, lb .45851
# weights applied 0.49772 with base_weight=0.01 and lower=0.06 and upper=0.7 lr: 0.15, msc:50, lb = .44151
```
